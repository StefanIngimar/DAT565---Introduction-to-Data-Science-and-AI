```python
import tarfile
from bs4 import BeautifulSoup
import pandas as pd
import os
import csv
from matplotlib import pyplot as plt
import numpy as np

with tarfile.open('kungalv_slutpriser.tar.gz', 'r:gz') as tar:
    tar.extractall('extracted_files', filter=None)  # Extracts to a folder named 'extracted_files'
```

```python
directory_path = 'extracted_files/kungalv_slutpriser'
data = []

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    
    # Check if it's a file
    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            contents = f.read()
            soup = BeautifulSoup(contents, "html.parser")
            
            # Find all listings in the file
            listings = soup.find_all("div", class_="hcl-padded-container hcl-flex--container hcl-flex--direction-column")
            
            print(f"Prices from {filename}:")
            for listing in listings:
                #Find the final prices and dates, and prepare them to be written into the csv file below
                row_data = {"price": None, "date": None}
                # Locate the span containing 'Slutpris' and the associated price
                price_span = listing.find("span", string=lambda text: text and "Slutpris" in text)
                if price_span:
                    row_data["price"] = price_span.get_text(strip=True).replace("Slutpris", "").strip()
                    
                date_element = soup.find("li", class_="sold-results__age-heading-item")
                if date_element:
                    row_data["date"] = date_element.find("span").get_text(strip=True)

                if row_data["price"] and row_data["date"]:
                    data.append(row_data)
                
                if price_span:
                    # Extract the price text, after 'Slutpris'
                    price_text = price_span.get_text(strip=True).replace("Slutpris", "").strip()
                    sold_date = price_span.get_text(strip=True).replace("Sålt i", "").strip()
                    print(price_text)

                if date_element:
                    # Extract the date text
                    sold_date = date_element.find("span").get_text(strip=True)
                    print(f"Sold Date: {sold_date}")
            print("\n" + "="*40 + "\n")
```

```python
#Following https://docs.python.org/3/library/csv.html
with open('extracted_files/price_data.csv', 'w', newline='') as csvfile:
    fieldnames = ['price', 'date']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerows(data)
```

```python
df = pd.read_csv('extracted_files/price_data.csv')
print(df.to_string())
```

```python
df['price'] = df['price'].str.replace(r'\D', '', regex=True).astype(int)

# Extract year from date column
df['year'] = df['date'].str.extract(r'(\d{4})')[0]

# Group by year and calculate the average price per year
df_2022 = df[df['year'] == "2022"]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_2022['date'], df_2022['price'], label="Price in 2022", marker='o')
plt.xlabel("Date")
plt.ylabel("Price (in kr)")
plt.title("Property Prices in 2022")
plt.legend()
plt.show()
```

```python
min_val = np.min(df_2022['price'])
q1 = np.percentile(df_2022['price'], 25)
median = np.median(df_2022['price'])
q3 = np.percentile(df_2022['price'], 75)
max_val = np.max(df_2022['price'])
print("Five-Number Summary:")
print(f"Minimum: {min_val}")
print(f"Q1: {q1}")
print(f"Median: {median}")
print(f"Q3: {q3}")
print(f"Maximum: {max_val}")
```
