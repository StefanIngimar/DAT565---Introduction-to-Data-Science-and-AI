```python
import torch
from PIL import Image
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
```

```python
sample_image, sample_label = train_dataset[0]
print("img shape:", sample_image.shape)
print("label:", sample_label)
print("pixel range:", sample_image.min().item(), "to", sample_image.max().item())
```

```python
import matplotlib.pyplot as plt

def plot_images(dataset, title):
    plt.figure(figsize=(10,4))
    for i in range(5):
        image, label = dataset[i]
        image = image.squeeze(0).numpy()
        plt.subplot(1, 5, i + 1)
        plt.imshow(image, cmap='gray')
        plt.title(f"label: {label}")
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

plot_images(train_dataset, "training set examples")
plot_images(test_dataset, "test set examples")
```
